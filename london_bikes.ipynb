```python
# Rename some columns for convenience

bikes_data = bikes_data\
    .rename(columns=
    {'timestamp': 'time',
    'cnt': 'count',
    't1': 'temp_real_C',
    't2': 'temp_feels_C',
    'hum': 'humidity_percent',
    'weather_code': 'weather',
    })

bikes_data.head()
```

```python
# Change the humidity values to percentage (value between 0 and 1)

bikes_data['humidity_percent'] = bikes_data['humidity_percent'] / 100

bikes_data.humidity_percent.head()
```

```python
# Create dictionaries of the seasons and the weather. So that I can map the integers to the actual written values

season_dict = {
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'autumn',
    '3.0': 'winter'
}

weather_dict = {
    '1.0': 'Clear',
    '2.0': 'Scattered clouds',
    '3.0': 'Broken clouds',
    '4.0': 'Cloudy',
    '7.0': 'Rain',
    '10.0': 'Rain with thunderstorm',
    '26.0': 'Snowfall',
}

# Changing the columns datatype to string

bikes_data.season = bikes_data.season.astype('str')
bikes_data.weather = bikes_data.weather.astype('str')

# Mapping the values to the actual written 

bikes_data.season = bikes_data.season.map(season_dict)
bikes_data.weather = bikes_data.weather.map(weather_dict)
```

```python
# Checking the dataframe after all these operations

bikes_data.head()
```

```python
# Export the final dataframe to an excel file

bikes_data.to_excel('bikes_data_final.xlsx', sheet_name='Data')
```
